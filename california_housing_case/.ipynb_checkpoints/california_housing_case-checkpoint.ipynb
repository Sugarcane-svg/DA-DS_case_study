```python
# split data into train and test set using self-defined function

import numpy as np

np.random.seed(10)
def train_test_split(data, ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_len = int(len(data) * ratio)
    test_indices = shuffled_indices[:test_len]
    train_indices = shuffled_indices[test_len:]
    return data.iloc[train_indices], data.iloc[test_indices]


# test the length of training and test set
train_set_1, test_set_1 = train_test_split(housing, 0.2)
print(str(len(train_set_1)))
print(str(len(test_set_1)))
```

```python
# use scikit-learn function to split data
from sklearn.model_selection import train_test_split
train_set_2, test_set_2 = train_test_split(housing, test_size = 0.2, random_state = 10)

print(str(len(train_set_2)))
print(str(len(test_set_2)))
```

```python
# prepare for machine learning algo
# 1. revert to clean training set
housing = strat_train_set.drop("median_house_value", axis = 1)
housing_label = strat_train_set["median_house_value"].copy()
```

```python
# 2. data cleaning: missing value by imputation with median value
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
housing_num = housing.drop("ocean_proximity", axis = 1) # imputer only work on numerical attribute, so remove categorical attribute first
imputer.fit(housing_num) # calculate median value for each attribute
X = imputer.transform(housing_num) # fill in median value when missing 
housing_tr = pd.DataFrame(X, columns = housing_num.columns) # convert resulting numpy array to pandas data frame
```

```python
# 3. handling categorical attribute
housing_cat = housing[["ocean_proximity"]]
housing_cat.head()
```

```python
# sklearn encoder to convert string into numbers 
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()
housing_cat_encoded = one_hot_encoder.fit_transform(housing_cat)
print(housing_cat_encoded) # this is sparse matrix instead of numpy array
print(housing_cat_encoded.toarray()) 
print(one_hot_encoder.categories_)
```

```python
# sometimes the default transformer is not enough for us to automate the process(i.e. custom cleanup operation 
# or combining specific attributes), we implement custom transformer

from sklearn.base import BaseEstimator, TransformerMixin 
# BaseEstimator: good for automate hyperparameter tuning
# TransformerMixin: get fit_transform() method for free

# column index in the dataset 
rooms_ix, bedrooms_ix, population_ix, households_ix = 3,4,5,6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room

    def fit(self, X, y = None):
        return self

    def transform(self, X, y = None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]

        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]

        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room = False)
housing_extra_attribs = attr_adder.transform(housing.values)
print(housing_extra_attribs)
```

```python
# transformation pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler # standardization less sensitive to outliers
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = "median")), # replace missing with median
    ('attribs_adder', CombinedAttributesAdder()), # custom transformer to get rooms_per_household and so on attributes
    ('std_scaler', StandardScaler()),
])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
# so far, the data has been splited into categorical and numerical, therefore, we handle the two data type separatly.
# sklearn provides a class which can handle them all in once
from sklearn.compose import ColumnTransformer
num_attribs = list(housing_num) # get the numeric column names
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs),
])

housing_prepared = full_pipeline.fit_transform(housing)
```

```python
# 4. train the model
# linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_label)

# try out some instances
some_data = housing.iloc[:5]
some_label = housing_label.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("prediction:", lin_reg.predict(some_data_prepared))
print("label:", list(some_label))

# measure RMSE
from sklearn.metrics import mean_squared_error
housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_label, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse) 

# the rmse is pretty large despite the fact that the median_house_value range between 120,000 and 265,000.
# so the model seems underfit the training set :(
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared, housing_label)

# measure rmse
housing_predictions = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_label, housing_predictions)
tree_rmse = np.sqrt(tree_mse)
print(tree_rmse) # 0 error! we need cross validation

from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_label, scoring = "neg_mean_squared_error", cv = 10)
tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores: ", scores)
    print("Mean: ",scores.mean())
    print("Std: ", scores.std())

print("Decision Tree RMSE score: \n", display_scores(tree_rmse_scores))
#display_scores(tree_rmse_scores) # seems worse than linear regression

# compute cv for linear regression
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_label, scoring = "neg_mean_squared_error", cv = 10)
lin_rmse_score = np.sqrt(-lin_scores)
print("Linear Regression RMSE score: \n", display_scores(lin_rmse_score))
#display_scores(lin_rmse_score)

# DecisionTree overfitting the training set, it performs worse than linear regression
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(housing_prepared, housing_label)

housing_predictions = rf_reg.predict(housing_prepared)
rf_mse = mean_squared_error(housing_label, housing_predictions)
rf_rmse = np.sqrt(rf_mse) # training set rmse value
print(rf_rmse)

rf_scores = cross_val_score(rf_reg, housing_prepared, housing_label, scoring = "neg_mean_squared_error", cv = 10)
rf_rmse_score = np.sqrt(-rf_scores)
print("RandomForest RMSE score: \n", display_scores(rf_rmse_score))
#display_scores(rf_rmse_score) # validation rmse value

# it seems training rmse is much lower than the validation rmse, therefore, the model is overfitting the training set

# ways to solve overfit: simplify model, constrain model(regularization), or get more training data
```

```python
# 5. tune your model
# GridSearchCV: when search space is small
# RandomizedSearchCV: when search space is large
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2,4,6,8]}, # 3 * 4 = 12 combinations
    {'bootstrap': [False], 'n_estimators':[3, 10], 'max_features':[2,3,4]}, # 6 combinations
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv = 5, scoring = "neg_mean_squared_error", return_train_score = True)

grid_search.fit(housing_prepared, housing_label)

print(grid_search.best_params_) # get the best paramters
print(grid_search.best_estimator_) # get the best estimators

# print out all evaluation scores
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
# 6. analyze the best model and their errors
feature_importance = grid_search.best_estimator_.feature_importances_
print("feature importance score: ", feature_importance)

# display the importance score with their attributes
extra_attribs = ["rooms_per_household", "population_per_household", "bedrooms_per_room"] # add the name if the attributes
cat_encoder = full_pipeline.named_transformers_["cat"] # get the one_hot_encoder
cat_one_hot_attribs = list(cat_encoder.categories_[0]) # use one_hot_encoder to get the encoded attributes
attributes = num_attribs + extra_attribs + cat_one_hot_attribs # combine attributes' names together
print(sorted(zip(feature_importance, attributes), reverse = True))
```

```python
# 7. evaluate your system on the test set
final_model = grid_search.best_estimator_
X_test = strat_test_set.drop("median_house_value", axis = 1)
y_test = strat_test_set["median_house_value"].copy()

X_test_prepared = full_pipeline.transform(X_test) # do not call fit_transform because you do not want to fit the test set
final_prediction = final_model.predict(X_test_prepared)

final_mse = mean_squared_error(y_test, final_prediction)
final_rmse = np.sqrt(final_mse)
print(final_rmse)

# see how good is our final_rmse
from scipy import stats
confidence = 0.95
squared_error = (final_prediction - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_error)-1, loc = squared_error.mean(), scale = stats.sem(squared_error)))
```

```python
# try SVM and see how it goes
from sklearn.svm import SVR

svr_reg = SVR(kernel = "linear")
svr_reg.fit(housing_prepared, housing_label)

housing_prediction = svr_reg.predict(housing_prepared)

svr_mse = mean_squared_error(housing_label, housing_prediction)
svr_rmse = np.sqrt(svr_mse)
print(svr_rmse)


svr_reg = SVR(kernel = "poly")
svr_reg.fit(housing_prepared, housing_label)

housing_prediction = svr_reg.predict(housing_prepared)

svr_mse = mean_squared_error(housing_label, housing_prediction)
svr_rmse = np.sqrt(svr_mse)
print(svr_rmse)

svr_reg = SVR(kernel = "rbf")
svr_reg.fit(housing_prepared, housing_label)

housing_prediction = svr_reg.predict(housing_prepared)

svr_mse = mean_squared_error(housing_label, housing_prediction)
svr_rmse = np.sqrt(svr_mse)
print(svr_rmse)
```

```python
# use RandomSearchCV and see if the performance improves
from sklearn.model_selection import RandomizedSearchCV

param_distribution = {
    "gamma": [1, 0.1, 0.01, 0.001],
    "kernel": ["linear", "poly", "rbf", "sigmoid"],
    "C": [0.1, 1, 10, 100]
}

model = SVR()
random_search = RandomizedSearchCV(
    estimator = model,
    cv = 5,
    scoring = "neg_mean_squared_error",
    param_distributions = param_distribution,
    # n_jobs = -1, only works in jupyter notebook
)

random_search.fit(housing_prepared, housing_label)

print("best params: ", random_search.best_params_)
print("best score: ", np.sqrt(-random_search.best_score_))
```
